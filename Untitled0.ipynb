```python
df.groupby('Brand', as_index = False).agg({'Price_USD':'mean'}).sort_values('Price_USD', ascending = False)
```

```python
plt.figure(figsize=(14, 6))
brand_prices = df.groupby('Brand')['Price_USD'].mean().sort_values(ascending = False)

# Рассчитываем оптимальный размер
n_brands = len(brand_prices)
rotation_angle = 90 if n_brands > 30 else 45

brand_prices.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Марки автомобилей')
plt.ylabel('Средняя стоимость (USD)')
plt.title(f'Средняя стоимость по {n_brands} маркам')
plt.xticks(rotation=rotation_angle, ha='right' if rotation_angle == 45 else 'center')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
df.loc[df['Model_Year']==2024] \
.groupby('Brand')\
.agg({'Price_USD':'mean'}) \
.sort_values('Price_USD')
```

```python
df_group_trans = df.loc[df['Model_Year']==2024].groupby(['Transmission', 'Brand']).agg({'Price_USD':'mean'})

plt.figure(figsize=(14, 6))
sns.barplot(x="Brand", y="Price_USD", data=df_group_trans, hue='Transmission', palette='Set2')
```
